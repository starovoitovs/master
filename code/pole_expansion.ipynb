Pole expansion of the kernel

```python
%matplotlib inline

import numpy as np
from scipy.special import gamma
import matplotlib.pyplot as plt

alpha = 0.6

@np.vectorize
def f1(z):
    return gamma(1-z) / gamma(1+alpha-z)

@np.vectorize
def f2(z, n=170):
    return - np.sum([(-1) ** k * 1 / gamma(alpha - k) / gamma(k + 1) / (z - (k + 1)) for k in range(0, n)])

x = np.linspace(0, 0.5, 101)

plt.plot(x, f1(x))
plt.plot(x, f2(x))

plt.show()
```
